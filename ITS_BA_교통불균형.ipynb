```python
import pandas as pd

pop_change=pd.read_csv('seoul_daily_pop.csv',encoding='cp949')
#pop_change
#월별 각 자치구에 있는 자치구간이동인구수/총 생활인구수

pop_change['기준일ID']=pd.to_datetime(pop_change['기준일ID'],format='%Y%m%d')

#10-12월은 2022년, 1-9월은 2023년
pop_change=pop_change[(pop_change['기준일ID']>='2022-10-01') & (pop_change['기준일ID']<='2023-09-30')]


pop_change['year']=pop_change['기준일ID'].dt.year
pop_change['month']=pop_change['기준일ID'].dt.month

#각 자치구에 있는 자치구간이동인구수/총 생활인구수
pop_change['이동인구비율']=pop_change['자치구간이동인구수']/pop_change['총생활인구수']
#같은 자치구에 대해서 그룹화
gu_by_month=pop_change.groupby(['시군구명','month'])

gu_by_month=gu_by_month['이동인구비율'].mean().reset_index()
gu_by_month.to_csv('gu_by_month.csv')

#매달 자치구별 이동인구비율의 평균을 취해서 해의 이동인구비율을 구해라
gu_year=gu_by_month.groupby('시군구명')['이동인구비율'].mean().reset_index()
gu_year=gu_year.sort_values(by='이동인구비율',ascending=False)
gu_year = gu_year[gu_year['시군구명'] != '서울시']

gu_year = gu_year.reset_index(drop=True)

gu_year.to_csv('자치구별이동인구비율.csv')

from google.colab import files
files.download('자치구별이동인구비율.csv')
files.download('gu_by_month.csv')



```

```python
#minmax scaler 적용
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

scaler=MinMaxScaler()

pop_change=pd.read_csv('seoul_daily_pop.csv',encoding='cp949')
#pop_change
#월별 각 자치구에 있는 자치구간이동인구수/총 생활인구수

pop_change['기준일ID']=pd.to_datetime(pop_change['기준일ID'],format='%Y%m%d')

#10-12월은 2022년, 1-9월은 2023년
pop_change=pop_change[(pop_change['기준일ID']>='2022-10-01') & (pop_change['기준일ID']<='2023-09-30')]


pop_change['year']=pop_change['기준일ID'].dt.year
pop_change['month']=pop_change['기준일ID'].dt.month

#정규화
data_to_normalize=pop_change[['자치구간이동인구수','총생활인구수']]
normalized_data = scaler.fit_transform(data_to_normalize)
pop_change[['자치구간이동인구수', '총생활인구수']] = normalized_data
print(pop_change[['자치구간이동인구수', '총생활인구수']])

#각 자치구에 있는 자치구간이동인구수/총 생활인구수
pop_change['이동인구비율']=pop_change['자치구간이동인구수']/pop_change['총생활인구수']
#같은 자치구에 대해서 그룹화
print(pop_change['이동인구비율'])
gu_by_month=pop_change.groupby(['시군구명','month'])

gu_by_month=gu_by_month['이동인구비율'].mean().reset_index()
gu_by_month.to_csv('normalized_gu_by_month.csv')

#매달 자치구별 이동인구비율의 평균을 취해서 해의 이동인구비율을 구해라
gu_year=gu_by_month.groupby('시군구명')['이동인구비율'].mean().reset_index()
gu_year=gu_year.sort_values(by='이동인구비율',ascending=False)
gu_year = gu_year[gu_year['시군구명'] != '서울시']

gu_year = gu_year.reset_index(drop=True)

gu_year.to_csv('정규화된자치구별이동인구비율.csv')

from google.colab import files
#files.download('정규화된자치구별이동인구비율.csv')
#files.download('normalized_gu_by_month.csv')
```
